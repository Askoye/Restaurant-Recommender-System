```
# check missing data
test.isnull().values.any()
```

```
# drop time1 and time2 in train
cols = range(33, 61)

test.drop(test.columns[[cols]], axis=1, inplace=True)
test.drop(test.columns[[2,10]], axis=1, inplace=True)
test.shape
```

```
# Renaming values in primary_tags
test['primary_tags'] = test['primary_tags'].str.replace(r'\D', '')
test.head()
```

```
# check % of null values
pd.DataFrame(test.isnull().mean() * 100).reset_index()
```

```
# drop commission, open_close_flags, language_x and language_y. Also dropped duplicated rows customer_id, id_obj and location number_obj

test.drop(test.columns[[4,7,23,24,29,34,35,36,37,41,42,43]], axis=1, inplace=True)
test.shape
```

```
# check % of null values
pd.DataFrame(test.isnull().mean() * 100).reset_index()
```

```
# check unique values in gender
#
test['gender'].unique()
```

```
# Replace values to male, female, and unknown
#
test['gender'] = test['gender'].replace(['Male', 'Male  ', 'male'], 'Male')
test['gender'] = test['gender'].replace(['Female', 'Female  ', 'Female   ', 'Female    ', 'Female '], 'Female')
test['gender'] = test['gender'].replace(['  ','?????'], 'Unknown')

# Filling null values Gender column
#
test['gender'] = test['gender'].fillna('Unknown')

# check_unique values
test['gender'].unique()
```

```
# Replacing na values in vendor_tag with 0 
test["vendor_tag"].fillna("0", inplace = True) 

test.head() 
```

```
# replace unkown tags with 0
test['primary_tags'].fillna('0', inplace=True)
```

```
# replace missing values with dash
#
test['OpeningTime'] = test['OpeningTime'].fillna('08:00AM-11:00PM')
test['OpeningTime2'] = test['OpeningTime2'].fillna('12:00AM-12:00AM')

test['OpeningTime'] = test['OpeningTime'].replace('-', '08:00AM-11:00PM')
test['OpeningTime2'] = test['OpeningTime2'].replace('-', '12:00AM-12:00AM')
```

```
# check number of vendors is accurate
len(test['id'].unique())
```

```
# check unique values for anomalies
test['OpeningTime'].unique()
```

```
# sort anomalies in the records
#
test['OpeningTime'] = test['OpeningTime'].str.replace('08:00AM-11:45-', '08:00AM-11:45PM')
test['OpeningTime'] = test['OpeningTime'].str.replace('9-22', '09:00AM-10:00PM')
test['OpeningTime'] = test['OpeningTime'].str.replace(' ', '') # remove leading and trailing white spaces
test['OpeningTime'] = test['OpeningTime'].str.replace('am','AM')
test['OpeningTime'] = test['OpeningTime'].str.replace('pm','PM')
test['OpeningTime'] = test['OpeningTime'].str.replace('9AM-10PM', '09:00AM-10:00PM')
test['OpeningTime'] = test['OpeningTime'].str.replace('11AM-11PM', '11:00AM-11:00PM')
test['OpeningTime'] = test['OpeningTime'].str.replace('01.00PM-02:.00AM', '01:00PM-02:00AM')
test['OpeningTime'] = test['OpeningTime'].str.replace('11:00AM-111:00PM', '11:00AM-11:00PM')
test['OpeningTime'] = test['OpeningTime'].str.replace('8:00AM-09.45PM', '08:00AM-09:45PM')
test['OpeningTime'] = test['OpeningTime'].str.replace('10:59AM-3:30PM', '10:59AM-03:30PM')
test['OpeningTime'] = test['OpeningTime'].str.replace('08:00AM-3:30PM', '08:00AM-03:30PM')
test['OpeningTime'] = test['OpeningTime'].str.replace('10:00AM-1:45PM', '10:00AM-01:45PM')
test['OpeningTime'] = test['OpeningTime'].str.replace('9:00AM-11:00PM', '09:00AM-11:00PM')
test['OpeningTime'] = test['OpeningTime'].str.replace('9:00AM-11:00PM', '09:00AM-11:00PM')
test['OpeningTime'] = test['OpeningTime'].str.replace('11:59AM-2:15AM', '11:59AM-02:15AM')
test['OpeningTime'] = test['OpeningTime'].str.replace('11:00AM-3:50PM', '11:00AM-03:50PM')
test['OpeningTime'] = test['OpeningTime'].str.replace('7:58AM-11:45PM', '07:58AM-11:45PM')
test['OpeningTime'] = test['OpeningTime'].str.replace('009:00AM-11:00PM', '09:00AM-11:00PM')
test['OpeningTime'] = test['OpeningTime'].str.replace('00:01AM-01:00AM', '12:01AM-01:00AM')
test['OpeningTime'] = test['OpeningTime'].str.replace('.', ':')
```

```
# split openingtime columns
#
test[['Open', 'Close']] = test.OpeningTime.str.split('-', expand=True, n=2)
test[['Open2', 'Close2']] = test.OpeningTime2.str.split('-', expand=True, n=2)
test.head()
```

```
# drop OpeningTime columns
test.drop(test.columns[[18, 19]], axis=1, inplace=True)
test.shape
```

```
# convert string time to time format
import datetime

open, close, open2, close2  = test.Open, test.Close, test.Open2, test.Close2
open_time, close_time, open2_time, close2_time = [], [] ,[], []

for time in open:
  # time = timeconvert(time)
  time = datetime.datetime.strptime(time, '%I:%M%p').time()
  open_time.append(time)

for time in close:
  # time = timeconvert(time)
  time = datetime.datetime.strptime(time, '%I:%M%p').time()
  close_time.append(time)

for time in open2:
  # time = timeconvert(time)
  time = datetime.datetime.strptime(time, '%I:%M%p').time()
  open2_time.append(time)

for time in close2:
  # time = timeconvert(time)
  time = datetime.datetime.strptime(time, '%I:%M%p').time()
  close2_time.append(time)

test.Open, test.Close, test.Open2, test.Close2 = open_time, close_time, open2_time, close2_time
test.head()
```

```
test.shape
```

```
# split string
test.Open = test.Open.astype(str).str.split(':').str[0]
test.Close = test.Close.astype(str).str.split(':').str[0]
test.Open2 = test.Open2.astype(str).str.split(':').str[0]
test.Close2 = test.Close2.astype(str).str.split(':').str[0]
test.head()
```

```
# compute hours vendor is open
test['duration'] = (((test.Close).astype(int) - (test.Open).astype(int)) + ((test.Close2).astype(int) - (test.Open2).astype(int)))
test.head()
```
