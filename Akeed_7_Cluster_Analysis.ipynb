## Loading Required Libraries and Data Exploration

```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)
sns.set_style('darkgrid')
```

```
clustered = pd.read_csv('/content/drive/My Drive/Akeed Datasets/clustered_train.csv')
```

```
# checking shape
clustered.shape
```

```
clustered['target'].value_counts()
```

```
clustered.head(2)
```

```
# renaming prepration column correctly, verified x and y, duration etc

clustered = clustered.rename(columns = {"prepration_time": "preparation_time", "verified_x": "customer_acc_verified", 'status_x': 'customer_status',
                                                "verified_y": "vendor_acc_verified", 'status_y': 'vendor_status',
                                                 "duration": "vendor_duration_open" })
```

```
# confirming that the columns have been renamed properly
clustered.head(2)
```

Yes, the column have been renamed correctly.

```
class1 = clustered[clustered['target'] == 1]
class1.shape
```

```
class1.head(2)
```

## Bivariate Analysis

Normalized K-Means Clusters with Target 

```
# converting normalized kmeans to category
clustered['normalized_kmeans'] = clustered['normalized_kmeans'].astype('category')
```

```
sns.catplot(x="target", hue = "normalized_kmeans", kind="count", data=clustered, aspect = 2)
plt.xlabel('Target', fontsize = 14, fontweight = 'bold')
plt.ylabel('Total Count', fontsize = 14, fontweight = 'bold')
plt.xticks(fontsize = 15, fontweight = 'bold')
plt.yticks(fontsize = 15, fontweight = 'bold')
plt.title("Comparing Normalized K-Means Clusters with Target", fontsize = 14, fontweight = 'bold')
```

People who made the orders are from clusters 2, 1 and 0.
